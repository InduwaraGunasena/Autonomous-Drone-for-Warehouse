# **Process JSON to Mask**

We need a data set to train our model. To create a data set we need to annotate all the image we have. There are lots of softwares, tools but I use free web base tool called [ImgLab](https://solothought.com/imglab/).

## **Import neccesory libraries**

```python
import numpy as np
import cv2
import json
```

# **Reading the JSON file**

```python
json_file = open("/content/drive/MyDrive/Colab Notebooks/mask5.json", 'r')
data = json.load(json_file)
print(data)
```

# **Folders**

```python
img_dir = "/content/drive/MyDrive/Colab Notebooks/clean images"
mask_dir = "/content/drive/MyDrive/Colab Notebooks/masks"
```

## **Images and Annotation**

```python
images = data['images']
annotations = data['annotations']
```

```python
print(images)
```

```python
print(annotations)
```

## **Loop over the data and save the mask**

I consider multiple masks on same image. So don't worry about how many masks it has. üòè

```python
for x in images:
    # Extract image details
    filename = x['file_name']
    height = x["height"]
    width = x["width"]
    img_id = x["id"]

    # Create an empty mask for the image
    mask = np.zeros((height, width))

    # Loop through all annotations
    for y in annotations:
        # Check if the annotation belongs to the current image
        if y["image_id"] == img_id:
            # Extract segmentation points
            seg = y["segmentation"]

            # Draw each segmentation contour on the mask
            for points in seg:
                contours = []

                # Convert the segmentation points into contours
                for i in range(0, len(points), 2):
                    contours.append([points[i], points[i + 1]])

                # Convert contours to a numpy array
                contours = np.array(contours, dtype=np.int32)

                # Draw the contours on the mask
                cv2.drawContours(mask, [contours], -1, 255, -1)

    # Save the final mask once all contours for the image are drawn
    cv2.imwrite(f"{mask_dir}/{filename}", mask)
```
